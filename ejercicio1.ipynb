## **Series crear y operar**

```python
#se crea la serie, usando pd.Series, que es un metodo de pandas. Le agregamos en array los datos que tendra la serie
Serie = pd.Series([10, 20, 30, 40, 50])
#se muestra
print(Serie)
#para acceder a un valor, lo hacemos de la misma forma que con una lista.
#accedemos llamando a la serie y entre corchetes la posicion que queremos editar
#lo igualamos con la misma posicion que se quiere editar y simplemente hacemos la operacion que queramos
Serie[1] = Serie[1] *2
#se muestra
print(Serie)
```

## **3) Inspeccionar y explorar**

```python
import pandas as pd

# Se crea el diccionario con los datos
dataframePequeño = {
    "nombre": ["richi", "juanes", "sebas"],
    "edad": [50, 18, 18],
    "curso": ["albañil", "ADSO", "ADSO"]
}

# Se convierte en DataFrame
dfp = pd.DataFrame(dataframePequeño)

#  Mostrar primeras filas
print("Primeras filas:")
print(dfp.head())

#  Mostrar últimas filas
print("\nÚltimas filas:")
print(dfp.tail())

#  Información general del DataFrame
print("\nInformación del DataFrame:")
dfp.info()

#  Dimensiones del DataFrame
print("\nDimensiones (filas, columnas):")
print(dfp.shape)

# Estadísticas descriptivas (solo para columnas numéricas)
print("\nEstadísticas descriptivas:")
print(dfp.describe())

#.head -> muestra la primera fila
#.tail -> muestra la ultima
#.info -> informacion general
#.shape -> muestra las dimensiones
#.describe -> resumen estadistico pero solo para columnas numericas.
```

## **Acceder a columnas y filas**

```python
#Obtén la columna Nombre como Series.

serie_nombre = dfp["nombre"]
print(serie_nombre)
#tambien se puede asi:
serie_conpunto = dfp.nombre
print(serie_conpunto)

#Obtén la fila con índice 1 usando loc y iloc.

dfp.loc[1] # devuelve al fila cuyo indice sea 1
dfp.iloc[1] # busca la posicion numerica, empezando desde cero
#siempre devuelve la segunda fila del Dataframe, sin importar el valor del indice.

#Accede al valor específico (fila 1, columna Edad).
print("Accede al valor específico (fila 1, columna Edad)")
#1: busca la fila que le hayamos puesto, en este caso, buscara la columna 1
#2:luego, busca la columna por su nombre, en este caso "edad"
print(dfp.loc[1, "edad"])

# o tambien se puede de esta forma:
#1: busca la fila numero 1 ( osea la segunda fila, porque empieza desde cero 0)
#2: luego, busca la columna numero 1 (tambien la segunda columna por la misma razon)

print(dfp.iloc[1,1])


```

## **Operaciones básicas**

```python
# se crea una variable, se le asigna la columna edad, pero sumada en 5
aumentoen5 = dfp["edad"] + 5
# se muestra el resultado
print(aumentoen5)
# ahora, de esta sencilla forma se crea una nueva columna, agregamos 3 valores
dfp["Ingreso_mensual"] = [-100, 3600000, 3600000]
# se crea una nueva columna, pero se usa la anterior multiplicada por 12 para el efecito anual
dfp["Ingreso *12"] = dfp["Ingreso_mensual"] * 12
# se muestra el data frame actualizado.
print(dfp)
```

## **Filtrar datos**

```python
# simplemente, se obtiene la columna de edad y con una operacion de comparacion
# se filtran las edades mayores a 30
print(dfp["edad"] > 30)
# lo mismo para filtrar nombres, el resultado aparece en true, donde se halla encontrado la condicion
print(dfp["nombre"] == "richi")
```

## **Manejo de valores faltantes**

```python
print("¿Hay nulos?")
print(dfp.isna())  # Muestra un mapa de nulos (True/False)
print("\nCantidad de nulos por columna:")
print(dfp.isna().sum())  # Cuenta cuántos nulos hay por columna

# Edad: reemplazar nulos con 0
dfp["edad"] = dfp["edad"].fillna(0)

# Ciudad: si no existe, primero la creamos con algunos nulos
if "Ciudad" not in dfp.columns:
    dfp["Ciudad"] = [None, "Medellín", None]  # ejemplo

dfp["Ciudad"] = dfp["Ciudad"].fillna("Desconocido")

# Ingreso: si no existe, la creamos con algunos nulos
if "Ingreso" not in dfp.columns:
    dfp["Ingreso"] = [1000, None, 1500]  # ejemplo

mediana_ingreso = dfp["Ingreso"].median()
dfp["Ingreso"] = dfp["Ingreso"].fillna(mediana_ingreso)

print("\nDataFrame actualizado:")
print(dfp)
```

## **Ejercicio integrado**

```python
import pandas as pd

# Leer el CSV original
df = pd.read_csv("actividad_semana5.csv")

#Incrementar Edad en 5
df["Edad"] = df["Edad"].fillna(0) + 5

# Crear columna Ingreso_anual
df["Ingreso_anual"] = df["Ingreso"].fillna(0) * 12

#Filtrar personas con Ingreso_anual > 36000
df_filtrado = df[df["Ingreso_anual"] > 36000]

#Ordenar por Ingreso_anual descendente
df_ordenado = df_filtrado.sort_values(by="Ingreso_anual", ascending=False)

#Guardar como CSV
df_ordenado.to_csv("personas_filtradas.csv", index=False)

# Mostrar resultado final
print(df_ordenado)
```
